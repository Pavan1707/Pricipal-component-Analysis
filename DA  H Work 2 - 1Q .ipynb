#  Determine the minimum number of principal components that maintain
   95% and 99% of the variance.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy import stats
```

```python
df1 = pd.read_excel("C:\\Users\\pavan\\Downloads\\url_data.xlsx")
df1.head()
```

# As the first column contains the lables, it is dropped.

```python
df = df1.drop('data1',axis= 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
scaled_data
```

# PCA for maintaining 95% variance

```python
from sklearn.decomposition import PCA
pca_1 = PCA(0.95)
pca_1.fit(scaled_data)
z = pca_1.transform(scaled_data)
features = z.T
cov_matrix = np.cov(features)
print(cov_matrix)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)
eigen_values
```

```python
eigen_vectors
```

```python
print(z.shape)
```

# comment : Here, the number of components are redued to  22 for maintaining 95% variance.

# Explained variance ratio for pca_1

```python
exp_var1 = pca_1.explained_variance_ratio_
exp_var1
  
```

# Here the expected variance ratio is high for the first eigen value and it will reduce when the eigen values are changed.

# => PCA for maintaining 99% variance 

```python
from sklearn.decomposition import PCA
pca_2 = PCA(0.99)
pca_2.fit(scaled_data)
Y = pca_2.transform(scaled_data)
features = Y.T
cov_matrix1 = np.cov(features)
print(cov_matrix1)
```

```python
print(Y.shape)
```

```python
eigen_value1, eigen_vector1 = np.linalg.eig(cov_matrix1)
eigen_value1
```

```python
eigen_vector1
```
